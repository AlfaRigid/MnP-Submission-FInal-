```python
#ques 2b:

grade_points = {"AP": 10, "AA": 10, "AB": 9, "BB": 8, "BC": 7, "CC": 6}
student_data = {}
with open("student_records.csv", 'r') as file:
    header = file.readline().strip().split(',')
    for line in file:
        roll_number, department, credit, course_name, course_type, grade = line.strip().split(',')
        credit = int(credit)
        grade_point = grade_points[grade]
        
        if roll_number not in student_data:
            student_data[roll_number] = {"total_credits": 0, "weighted_points": 0}
            
        student_data[roll_number]["total_credits"] += credit
        student_data[roll_number]["weighted_points"] += credit * grade_point
        
for roll_number, info in student_data.items():
    total_credits = info["total_credits"]
    weighted_points = info["weighted_points"]
    cpi = weighted_points / total_credits if total_credits > 0 else 0
    print(f"Roll Number: {roll_number}, Total Credits: {total_credits}, CPI: {cpi:.2f}")
```

```python
#Ques 2c:
requirements = {"core": 20, "department_elective": 15, "flexible_elective": 10, "hasmed_elective": 5}

student_credits = {}
with open("student_records.csv", 'r') as file:
    header = file.readline().strip().split(',')
    for line in file:
        roll_number, department, credit, course_name, course_type, grade = line.strip().split(',')
        credit = int(credit)
        
        if roll_number not in student_credits:
            student_credits[roll_number] = {"core": 0, "department_elective": 0, "flexible_elective": 0, "hasmed_elective": 0}
        
        if course_type in student_credits[roll_number]:
            student_credits[roll_number][course_type] += credit

for roll_number, credits in student_credits.items():
    if all(credits[req] >= requirements[req] for req in requirements):
        print(f"Roll Number: {roll_number}")
```

```python
# Ques 2d:
minor_credits = {}
with open("student_records.csv", 'r') as file:
    header = file.readline().strip().split(',')
    for line in file:
        roll_number, department, credit, course_name, course_type, grade = line.strip().split(',')
        credit = int(credit)
        
        if course_type == "minor":
            if roll_number not in minor_credits:
                minor_credits[roll_number] = 0
            minor_credits[roll_number] += credit

for roll_number, total_credits in minor_credits.items():
    if total_credits >= 10:
        print(f"Roll Number: {roll_number}")
```

```python
#Ques 2e:
honours_credits = {}
core_credits = {}
with open("student_records.csv", 'r') as file:
    header = file.readline().strip().split(',')
    for line in file:
        roll_number, department, credit, course_name, course_type, grade = line.strip().split(',')
        credit = int(credit)
        
        if roll_number not in honours_credits:
            honours_credits[roll_number] = 0
            core_credits[roll_number] = 0
        
        if course_type == "honours":
            honours_credits[roll_number] += credit
        elif course_type == "core":
            core_credits[roll_number] += credit

for roll_number in honours_credits:
    if honours_credits[roll_number] >= 10 and core_credits[roll_number] >= 20:
        print(f"Roll Number: {roll_number}")
```

```python
# scipy PART A :
from scipy.optimize import minimize

def objective(vars):
    x, y = vars
    return 2 * (x - y - 3)**2 + 4 * (x + 2*y + 1)**4

def constraint1(vars):
    x, y = vars
    return x - y + 3  # x - y >= -3

def constraint2(vars):
    x, y = vars
    return 5 - ((x + 2)**2 + (y + 1)**2)  # (x + 2)^2 + (y + 1)^2 <= 5

initial_guess = [0, 0]

constraints = [
    {'type': 'ineq', 'fun': constraint1},  # inequality for constraint1
    {'type': 'ineq', 'fun': constraint2}   # inequality for constraint2
]

result = minimize(objective, initial_guess, constraints=constraints)

print("Optimal solution:", result.x)
print("Minimum value of f(x, y):", result.fun)
```

```python
# Scipy PART B 
from scipy.integrate import quad
import numpy as np

def f(x, y):
    return x**2 + y**4

R = np.sqrt(3)  # Radius of the circle

def x(t):
    return R * np.cos(t)

def y(t):
    return R * np.sin(t)

def dx_dt(t):
    return -R * np.sin(t)

def dy_dt(t):
    return R * np.cos(t)

def integrand(t):
    xi, yi = x(t), y(t)
    ds = np.sqrt(dx_dt(t)**2 + dy_dt(t)**2)
    return f(xi, yi) * ds

result, _ = quad(integrand, 0, 2 * np.pi)

print("Value of the line integral:", result)
```
